```python
%config Completer.use_jedi = False
import warnings
warnings.filterwarnings(action="ignore")

import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python

targetSite = "https://music.bugs.co.kr/chart?wl_ref=M_left_02_01/"
request = requests.get(targetSite)
html = request.text
soup = BeautifulSoup(html, "html.parser")
#print(soup)
```

```python
#노래 제목 크롤링
titles = soup.findAll("p", {"class":"title"})
#print(len(title))
#print(title)

for i in range(len(titles)):
    #print("{:3d}위, {}".format((i+1), titles[i].text.strip()))
    #print("{:3d}위, {}".format((i+1), titles[i].text.split("\n")[1]))
    print("{:3d}위, {}".format((i+1), titles[i].text.strip().split("\n")[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll("p", {"class":"artist"})
#print(len(artists))
#print(artists)

for i in range(len(artists)):
    #print("{:3d}위, {}".format((i+1), artists[i].text.strip())) # 적절하지 않음
    #print("{:3d}위, {}".format((i+1), artists[i].text.split("\n")[1])) #적절하지 않음
    print("{:3d}위, {}".format((i+1), artists[i].text.strip().split("\n")[0]))
```

```python
print(dt.now())
for i in range(len(titles)):
    artist = artists[i].text.strip().split("\n")[0]
    title = titles[i].text.strip().split("\n")[0]
    print("{:3d}위, {} - {}".format((i+1), artist, title))
```

```python
#크롤링한 결과를 텍스트 파일에 저장한다.
#file = open("C:\\Users\\tj\\Desktop\\python\\crawling\\output\\bugsTOP100.txt", "w") # 절대 경로
#file = open("C:/Users/tj/Desktop/python/crawling/output/bugsTOP100.txt", "w") # 절대 경로
file = open("./output/bugsTOP100_1.txt", "w") # 상대 경로
# 위의 open() 함수와 같이 파일을 열때 "cp949" 에러가 발생하면 아래와 같이 open한다.
#file = open("./output/bugsTOP100.txt", "w", -1, "utf-8")
#file.write(str(dt.now()) + "\n")
file.write(dt.now().strftime("%Y-%m-%d %H:%M:%S") + "\n")
for i in range(len(titles)):
    artist = artists[i].text.strip().split("\n")[0]
    title = titles[i].text.strip().split("\n")[0]
    file.write("{:3d}위, {} - {}\n".format((i+1), artist, title))

file.close()
print("bugsTOP100.txt 저장 됐다. 알겠냐?????앙???")
```

```python
#with 구문을 사용해서 파일을 open 하면 close() 메소드를 사용하지 않아도 with 블록의 내용을
#모두 실행하고 난 후 자동으로 파일을 닫는다.
with open("./output/bugsTOP100_2.txt", "w") as file :
    file.write(dt.now().strftime("%Y-%m-%d %H:%M:%S") + "\n")
    for i in range(len(titles)):
        artist = artists[i].text.strip().split("\n")[0]
        title = titles[i].text.strip().split("\n")[0]
        file.write("{:3d}위, {} - {}\n".format((i+1), artist, title))
    #-------- for
#------- with
print("bugsTOP100.txt 저장 됐다. 알겠냐?????앙???2")
```

```python
# 타겟 사이트 주소 만들기
#https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240417&hh=15&rtm=Y&pg=1 -> 1 - 50
#https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240417&hh=15&rtm=Y&pg=2 -> 51 -100
#https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240417&hh=15&rtm=Y&pg=3 -> 101 - 150
#https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240417&hh=15&rtm=Y&pg=4 -> 151 - 200

url = "https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240417&hh=15&rtm=Y&pg="
for i in range(1, 5):
    #targetSite = url + str(i)
    #targetSite = "{}{}".format(url,i)
    targetSite = f"{url}{i}"
    print(targetSite)
```

```python
#4개 페이지를 크롤링한 결과를 하나에 리스트에 모아준다
genie = []
url = "https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240417&hh=15&rtm=Y&pg="
h = {"User-agent" : "Mozilla/5.0 (Windows NT 6.3; Win64; x64; Trident/7.0; MSAppHost/2.0; rv:11.0) like Gecko"}
for i in range(1, 5):
    targetSite = url + str(i)
    request = requests.get(targetSite, headers=h)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html, "html.parser")
    
    titles = soup.findAll("a", {"class" : "title ellipsis"}) 
    artists = soup.findAll("a", {"class" : "artist ellipsis"})
    
    for j in range(len(titles)) :
        genie.append(
            ( artists[j + 5].text.strip().split("\n")[0], titles[j].text.strip().split("\n")[0] )
        )
    # ------ for j
# -------- for i
print(len(genie))
print(genie)
```

```python
print(dt.now())
for i in range(len(genie)):
     print("{:3d}위, {} - {}".format((i+1), genie[i][0], genie[i][1]))
```

```python
with open("./output/genieTOP200.txt", "w") as file :
    file.write(dt.now().strftime("%Y-%m-%d %H:%M:%S") + "\n")
    for i in range(len(genie)):
        file.write("{:3d}위, {} - {}\n".format((i+1), genie[i][0], genie[i][1]))
    #-------- for
#------- with
print("genieTOP200.txt 저장 됐다. 알겠냐?????앙???2")
```

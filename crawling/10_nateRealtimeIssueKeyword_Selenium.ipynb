```python
# -> 현재 사용중인 크롬 버전(124.0.6367.61)과 같은 버전의 드라이버를 os에 맞게 다운받는다.
#https://chromedriver.chromium.org/home -> 공식 사이트에는 114버전이 최상위 버전
#공식사이트의 the Chrome for Testing (CfT) availability dashboard 링크를 클릭하면 https://googlechromelabs.github.io/chrome-for-testing/ 페이지가
#실행되고 	https://storage.googleapis.com/chrome-for-testing-public/123.0.6312.122/win64/chromedriver-win64.zip를 브라우저에 입력해서 다운받는다.
# -> 123버전
#https://github.com/mozilla/geckodriver?tab=readme-ov-file -> 파이어폭스
#다운받은 크롬 드라이버를 워크스페이스에 복사해 넣고 사용한다.

#selenium을 사용해서 가상 크롬을 실행한다.

options = Options()
options.binary_location = r'C:\Program Files\Mozilla Firefox\firefox.exe'
driver = webdriver.Firefox( executable_path=r"C:/Users/tj/Desktop/python/crawling/geckodriver.exe", options=options)
#get() 메소드로 가상 크롬에 크롤링할 타겟 사이트를 띄운다
driver.get("https://www.nate.com")
#가상 크롬에 실행된 사이트의 html 정보를 얻어온다.
html = driver.page_source
#print(html)
soup = BeautifulSoup(html, "html.parser")
print(soup)
```

```python
issues = soup.findAll("a", {"class" : "ik"})
ranks = soup.select("#olLiveIssueKeyword .num_rank")
for i in range(5):
    rank = ranks[i].text
    issue = issues[i].text.strip().split("\n")[0]
    upDown = issues[i].text.strip().split("\n")[1]
    #print("%2s위 -> %s " % (rank, issue))
    print("{:>2s}위 -> {}".format(rank, issue), end="")
    if upDown[:2] == "상승":
        print("[{}{}]".format("↑", upDown[-1]))
    elif upDown[:2] == "하락":
        print("[{}{}]".format("↓", upDown[-1]))
    elif upDown[:2] == "동일":
        print("[-]")
    else :
        print("[new]")
```

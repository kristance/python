```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다.
import folium
```

```python
#folium의 Map 클래스의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 속성으로
#지정한 배율을 사용하는 지도를 만든다.
#folium.Map(location=[위도, 경도], zoom_start=배율)
folium_map = folium.Map(location=[37.569147, 126.984695], zoom_start=18)
folium_map
```

```python
folium_map = folium.Map(location=[37.569147, 126.984695], zoom_start=14)

#folium의 Marker 클래스의 location 속성으로 지정한 위치에 마커를 만든다.
# 마커는 add_to() 메소드로 지도에 추가한다.
#folium의 Marker 클래스의 popup 속성으로 마커를 클릭했을때 표시할 문자열을 지정할 수 있다.
folium.Marker(location=[37.57856527,126.97695439], popup='경복궁 근정전').add_to(folium_map)
folium.Marker(location=[37.56595852,126.97416479], popup='덕수궁 해시계').add_to(folium_map)
folium.Marker(location=[37.5787512,126.9948775], popup='창경궁').add_to(folium_map)
folium.Marker(location=[37.57943198,126.99104344], popup='창덕궁').add_to(folium_map)
folium.Marker(location=[37.57422213,126.99352068], popup='종묘').add_to(folium_map)
folium.Marker(location=[37.576034,126.98705201], popup='운현궁').add_to(folium_map)

# 마커를 클릭했을 때 팝업되는 말풍선 크기는 folium의 Popup 클래스의 max_width 속성으로 지정할 수 있다.
popup = folium.Popup('영풍문고 - 종로본점다아아아아아아', max_width=200)
folium.Marker(location=[37.56991173,126.98241074], popup=popup).add_to(folium_map)

#save() 메소드로 지도를 html 파일로 저장할 수 있다.
folium_map.save('./output/foliumMap.htm')
folium_map
```

```python
folium_map = folium.Map(location=[37.569147, 126.984695], zoom_start=14)

# folium의 Icon 클래스의 color 속성으로  마커의 색상과 icon 속성으로 아이콘 모양을 지정할 수 있다.
# 마커 색상과 아이콘 모양을 변경하려면 folium의 Marker 클래스의 icon 속성에 Icon 객체를 지정하면 된다.
# 색상 이름 리스트
# 'red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred',
#    'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white',
#     'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray'
# 마커에 표시할 아이콘은 bootstrap3에서 제공하는 glyphicons에서 찾을 수 있다. 가능유무는 실행해봐야암.
# https://getbootstrap.com/docs/3.3/components/

folium.Marker(location=[37.57856527,126.97695439], popup='경복궁 근정전',
            icon=folium.Icon(color='purple',icon='glass')).add_to(folium_map)
folium.Marker(location=[37.56595852,126.97416479], popup='덕수궁 해시계',
             icon=folium.Icon(color='cadetblue', icon='home')).add_to(folium_map)
folium.Marker(location=[37.5787512,126.9948775], popup='창경궁',
             icon=folium.Icon(color='lightred', icon='time')).add_to(folium_map)
folium.Marker(location=[37.57943198,126.99104344], popup='창덕궁',
             icon=folium.Icon(color='orange', icon='film')).add_to(folium_map)
folium.Marker(location=[37.57422213,126.99352068], popup='종묘',
             icon=folium.Icon(color='darkgreen', icon='remove')).add_to(folium_map)
folium.Marker(location=[37.576034,126.98705201], popup='운현궁',
             icon=folium.Icon(color='gray', icon='music')).add_to(folium_map)

# 마커를 클릭했을 때 팝업되는 말풍선 크기는 folium의 Popup 클래스의 max_width 속성으로 지정할 수 있다.
popup = folium.Popup('영풍문고 - 종로본점다아아아아아아', max_width=200)
folium.Marker(location=[37.56991173,126.98241074], popup=popup,
             icon=folium.Icon(color='red', icon='user')).add_to(folium_map)

#save() 메소드로 지도를 html 파일로 저장할 수 있다.
folium_map.save('./output/foliumMap2.htm')
folium_map
```

```python
folium_map = folium.Map(location=[37.569147, 126.984695], zoom_start=12)

#folium의 CircleMarker 클래스로 원형 마커를 만들 수 있다.
#radius 속성으로 마커로 표시되는 원의 반지를 지정한다.
#color 속성으로 원의 선 색상을, fill_color속성으로 원 내부에 채울 색상을 지정할 수 있다.

folium.CircleMarker(location=[37.56595852,126.97416479], popup='덕수궁 해시계',
                   radius=50, color='thistle', fill_color='purple').add_to(folium_map)

popup = folium.Popup('영풍문고 - 종로본점다아아아아아아', max_width=200)
folium.CircleMarker(location=[37.56991173,126.98241074], popup=popup,
                   radius=50, color='gray', fill_color='yellow').add_to(folium_map)

folium_map.save('./output/foliumMap3.htm')
folium_map
```

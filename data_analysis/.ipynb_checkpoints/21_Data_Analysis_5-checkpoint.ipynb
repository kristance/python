```python
# 집행장소가 누락값인 데이터가 존재하므로 누락값을 공백으로 채워준다.
df.집행장소.fillna('', inplace=True)
df[df.집행장소.isna()]
```

```python
# 집행장소가 '-'인 데이터가 존재하므로이 공백으로 변경한다.
df.loc[df.집행장소 == '-', '집행장소'] = ''
df[df.집행장소 == '-']
```

```python
# 집행장소가 '0'인 데이터가 존재하므로이 공백으로 변경한다.
df.loc[df.집행장소 == '0', '집행장소'] = ''
df[df.집행장소 == '0']
```

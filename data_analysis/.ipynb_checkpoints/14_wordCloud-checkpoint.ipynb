워드클라우드를 그레이스케일로 표시하는 함수를 만든다.


```python
import random
```

```python
# word : 워드클라우드로 표시할 단어
# font_size : 워드클라우드로 표시할 단어의 크기
# position : 워드클라우드로 표시할 단어의 위치(좌표)
# orientation : 워드클라우드로 표시할 단어의 방향
# random_state=None : 숫자를 넘겨주면 워드클라우드로 표시되는 단어들의 위치가 고정된다.
# **hwargs : 워드클라우드로 표시할 단어들의 글꼴이 저장된 경로
# *이 1개면 리스트 가변인자, *이 2개면 key:value 가변인자
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs):
    # print(word, font_size, position, orientation, random_state, hwargs)
    #return ('hsl(0, 0%%, %d%%)' % random.randint(30, 60)) # %%는 뒤에 % 중복때문에 사용
    return 'hsl(0, 0%, {}%)'.format( random.randint(0, 60) )
```

```python
plt.figure(figsize=[6,6])
wordcloud = WordCloud(max_font_size=100).generate(text)
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=200),
           interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 한글 형태소 분석을 위한 라이브러리를 import한다.
# pip install "jpypel<1"
# pip install konlpy
# pip install nltk
```

```python
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import nltk
okt = Okt()
```

```python
# konlpy가 제공하는 대한민국 헌법 말뭉치를 읽어온다.
kolaw_corpus = kolaw.open('constitution.txt').read()
print(kolaw_corpus[:40])
```

```python
# 한글 형태소 분석을 실행한다.
tokens_ko = okt.nouns(kolaw_corpus)
print(tokens_ko)
```

```python
#형태소 분석 결과 중에서 1음절 단어를 제거한다.
tokens_ko = [word for word in tokens_ko if len(word) > 1]
tokens_ko
```

```python
# 워드클라우드로 구현할 단어들의 출현 빈도수를 센다.
ko = nltk.Text(tokens_ko)
ko.vocab().most_common() # 출현 빈도수가 많은 단어부터 (단어, 갯수) 형태의 튜플로 출력한다.
```

```python
# 불용어를 제거한다.
stop_words = ['모든', '기타' '사항', '경우', '거나', '다만', '누구'] # 불용어를 기억하는 리스트
tokens_ko = [word for word in tokens_ko if word not in stop_words]
ko = nltk.Text(tokens_ko)
ko.vocab().most_common()
```

```python
#워드클라우드로 구현할 단어를 뽑아내서 단어를 key로 하고 출현 빈도수를 value로 하는 딕셔너를 만든다.
word = dict(ko.vocab().most_common(250))
word
```

```python
#딕셔너리에 저장된 데이터는
mask = np.array(Image.open('./image/korea.jpg'))
wordcloud = WordCloud(background_color='white',
                      max_font_size=50,
                      font_path= font,
                      mask=mask
                     ).generate_from_frequencies(word)
plt.figure(figsize=[6, 6])
plt.imshow(wordcloud, interpolation='spline36')
plt.axis('off')
plt.show()
```

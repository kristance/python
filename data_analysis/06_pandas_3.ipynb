***
concat() 메소드는 데이터프레임을 연결한다.  

연결하려는 데이터프레임을 리스트에 담아서 concat() 메소드에 전달하면 연결한 데이터프레임을 리턴한다.
***
- concat([데이터프레임1, 데이터프레임2, ....][, axis])  
- axis 속성을 생략하면 0이 기본값이고 열(위에서 아래) 방향으로 연결하고,  
    axis=1로 지정하면 행(왼쪽에서 오른쪽)방향으로 연결한다.

```python
# df1, df2, df3 데이터프레임은 열 인덱스가 모두 A, B, C, D로 같아서 연결한 다음에도 열이 그대로 유지된다.
row_concat = pd.concat([df1, df2, df3]) #row_concat = pd.concat([df1, df2, df3], axis=0), 위에서 아래방향으로 연결
row_concat
```

```python
# df1, df2, df3 데이터프레임은 행 인덱스가 모두 0, 1, 2, 3로 같아서 연결한 다음에도 행이 그대로 유지된다.
col_concat = pd.concat([df1, df2, df3], axis=1) # 왼쪽에서 오른쪽 방향으로 연결
col_concat
```

```python
# 같은 열 인덱스가 존재하는 데이터프레임에서 열 인덱스로 추출하면 해당 열 인덱스의 데이터가 모두 추출된다.
col_concat.A
```

***
데이터프레임에 데이터 1건 추가하기
***

```python
row_series = pd.Series(['n1', 'n2', 'n3', 'n4']) # 행 시리즈
row_series
```

```python
# axis를 생략해서 axis=0일때 행 시리즈가 새로운 행으로 추가되지 않고 새로운 열로 추가된다.
# 시리즈에 열 이름이 없기때문에 시리즈를 데이터프레임에 새로운 행으로 추가하려 할 때 제대로 추가되지 않는다.
# 추가하려는 시리즈를 새로운 열로 간주해서 0이라는 새로운 인덱스에 추가된다.
pd.concat([df1, row_series])
```

```python
#  pd.DataFrame(['n1', 'n2', 'n3', 'n4'])와 같이 1차원 리스트를 전달하면 데이터 하나하나를 개별 데이터를 취급해서
# 열 방향 데이터프레임이 생성되지만  [['n1', 'n2', 'n3', 'n4']] 와 같이 2차원 리스트로 전달하면 1차원 리스트 전체를
# 1개의 데이터로 취급해서 행 방향 데이터프레임이 생성된다.
row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']]) # 행 방향 데이터프레임이 생성된다.
row_df
```

```python
pd.concat([df1, row_df])
```

```python
# DataFrame()로 데이터프레임을 생성할 때 열 인덱스를 지정하지 않으면 pandas가 자동으로 0, 1, 2..와 같은 방식으로
# 열 인덱스를 지정하기 때문에 column 속성으로 열 인덱스를 전달한다.
row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D'])
row_df
```

```python
pd.concat([df1, row_df])
```

***
concat() 메소드는 여러개의 데이터프레임을 연결할 때 사용한다.  
만약에 데이터프레임을 딱 2개만 연결한다면 concat() 메소드도 가능하고, append() 메소드를 사용해도 된다.

```python
df1.append(row_df, ignore_index=True)
```

```python
pd.concat([df1, df2, df3], axis=1, ignore_index=True)
```

***
append() 메소드를 사용하면 데이터프레임을 사용하지 않고 딕셔너리를 사용해서 데이터프레임에  
추가할 수 있다.

```python
data_dict = {
    'A' : 'n5',
    'B' : 'n6',
    'C' : 'n7',
    'D' : 'n8'
}
print(data_dict)
#concat() 메소드의 인수는 시리즈나 데이터프레임만 지정할 수 있기때문에 에러가 발생된다.
# pd.concat([df1, data_dict]) # 에러
# 딕셔너리를 추가하려는 경우 딕셔너리에는 인덱스가 없으므로 ignore_index속성을 이용해서
# 추가 대상의 데이터프레임의 인덱스를 무시하게해야 딕셔너리를 추가할 수 있다.
#df1.append(data_dict) #에러
df1.append(data_dict, ignore_index=True)
```

***

***
concat() 메소드의 join 속성을 사용하면 연결하는 데이터프레임에 공통적으로 포함된 열만 골라서 연결할 수 있다. 


```python
# join 속성의 기본값은 outer이고 공통으로 데이터가 있는 열에 신경쓰지 않고 연결한다 -> NaN이 대량 발생한다.
pd.concat([df1, df2, df3], ignore_index=True, join='outer')
```

```python
# df1, df2, df3 데이터프레임은 3개 모두 공통 열 인덱스가 없기때문에 빈 데이터프레임이 생성된다.
pd.concat([df1, df2, df3], ignore_index=True, join='inner')
```

```python
# df1, df3 데이터프레임은 2개 모두 공통 열 인덱스가 있기때문에 공통 열을 연결한 데이터프레임이 생성된다.
pd.concat([df1, df3], ignore_index=True, join='inner')
```

```python
# df2, df3 데이터프레임은 2개 모두 공통 열 인덱스가 있기때문에 공통 열을 연결한 데이터프레임이 생성된다.
pd.concat([df2, df3], ignore_index=True, join='inner')
```

***
merge() 메소드로 2개의 데이터프레임을 병합한다.   
기본값으로 내부 조인을 실행하며, 메소드를 실행한 데이터프레임이 left로 지정되고  
메소드의 인수로 지정한 데이터프레임이 right로 지정된다.  
- left_on 속성과 right_on 속성은 각각 왼쪽과 오른쪽 데이터프레임에서 일치해야 할 열 인덱스를 지정하고,  
    how 속성으로 두 데이터프레임을 병합하는 방법을 지정한다.  
- 왼쪽데이터프레임.merge(오른쪽데이터프레임, left_on, right_on[, how])
***

```python
# survey_person.csv 파일과 survey_survey.csv 파일 merge
print(person.columns)
print(survey.columns)
person.merge(survey, left_on='ident', right_on='person', how='inner')
```

```python
print(person.columns)
print(survey.columns)
person.merge(survey, left_on='ident', right_on='person', how='outer')
```

```python
person.merge(survey, left_on='ident', right_on='person', how='left')
```

```python
p2s = person.merge(survey, left_on='ident', right_on='person', how='inner')
p2s
```

```python
# survey_visited.csv과 survey_survey.csv 파일 merge()
print(visited.columns)
print(survey.columns)
v2s = visited.merge(survey, left_on='ident', right_on='taken', how='inner')
v2s
```

```python
#p2s 데이터프레임과 v2s 데이터프레임 merge
print(p2s.columns)
print(v2s.columns)

# left_on 속성과 right_on 속성에 리스트로 묶어서 여러개의 열 인덱스를 전달할 수 있다.
p2s.merge(v2s, left_on=['ident', 'taken', 'quant', 'reading'], right_on=['person', 'taken', 'quant', 'reading'], how='inner')

# merge() 메소드를 실행하는 두 데이터프레임에 공통된 열 인덱스가 있으면 왼쪽 데이터프레임의 열 인덱스에는
#'_x'라는 접미사가 자동으로 추가되고, 오른쪽 데이터프레임의 열 인덱스에는 '_y'라는 접미사가 자동으로 추가된다.
```

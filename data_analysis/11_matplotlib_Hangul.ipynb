```python
%config Completer.use_jedi = False
import warnings
warnings.filterwarnings(action="ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
import matplotlib.font_manager as fm
# 한글 폰트를 사용할 때 마이너스 부호가 깨져보이는 문제에 대처한다.
mpl.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100, 100, 50).cumsum()
print(data)
```

```python
#plt.figure(figsize=[8, 6]) # 차트 크기 설정
plt.rcParams['figure.figsize'] = [8, 6] # 차트 크기 설정
plt.plot(range(1, 101, 2), data, '*-r')
plt.title('시간별 가격 추이') # 차트 제목
plt.xlabel('시간') # x축 제목
plt.ylabel('가격') # y축 제목
plt.show()
```

```python
# 1. FontProperties를 사용하는 방법 -> 그래프에 폰트가 필욯나 항목마다 일일이 개별적으로 설정한다.
# matplotlib.pyplot -> title(), xlabel(), ylabel(), legend(), text()
# matplotlib -> set_title()
# FontProperties(사용할 글꼴 경로, 글꼴 크기)
font = 'C:\\Windows\\Fonts\\NanumGothicCoding.ttf'
fontprop = fm.FontProperties(fname=font, size=16)
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
plt.plot(range(1, 101, 2), data, '*-r')
plt.title('시간별 가격 추이', fontproperties=fontprop) 
plt.xlabel('시간', fontproperties=fontprop) 
plt.ylabel('가격', fontproperties=fontprop)
plt.show()
```

```python
# 2. matplotlib 라이브러리 rcParams[]로 전역 글꼴 설정 -> 그래프에 폰트 설정이 필요한 파일마다 설정한다.
#matplotlib 폰트 기본 설정 정보 읽기
print('기본으로 설정된 폰트 이름 -> {}'.format(plt.rcParams('font.family')))
print('기본으로 설정된 폰트 크기 -> {}'.format(plt.rcParams('font.size')))
```

```python
# 2-1 
plt.rcParams('font.family') = 'NanumGothicCoding'
plt.rcParams('font.size') = 16
print('변경된 폰트 이름 -> {}'.format(plt.rcParams('font.family')))
print('변경된 폰트 크기 -> {}'.format(plt.rcParams('font.size')))
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
plt.plot(range(1, 101, 2), data, '*-r')
plt.title('시간별 가격 추이') 
plt.xlabel('시간') 
plt.ylabel('가격')
plt.show()
```

```python
#2-1
#rcParamseotls 대신 fontproperties를 이용해서 plt.rc()를 사용하는 방법이 있다.
font = 'C:\\Windows\\Fonts\\NanumGothicCoding.ttf'
fontprop = fm.FontProperties(fname=font, size=16).get_name()
plt.rc('font', family = fontprop)
plt.rc('font', size = 16)
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
plt.plot(range(1, 101, 2), data, '*-r')
plt.title('시간별 가격 추이') 
plt.xlabel('시간') 
plt.ylabel('가격')
plt.show()
```

```python
# 3. 폰트 설정 정보를 matplotlib 설정 팡리에 직접 넣어준다. -> 그래프에 폰트 설정이 필요한 컴퓨터마다 설정한다.
# mpl.matplotlib_fname()를 실행하면 출력되는 경로의 matplotlibrc 파일을 수정한다.
# matplotlibrc 파일을 수정하면 jupyter notebook을 실행할 때마다 별도의 설정없이 matplotlibrc 파일의
# 설정을 사용한다.
# font.family : san-serif -> NanumGothicCoding
# font.size : 10.0 -> 16.0
```

```python
plt.rcParams['figure.figsize'] = [8, 6]
plt.plot(range(1, 101, 2), data, '*-r')
plt.title('시간별 가격 추이') 
plt.xlabel('시간') 
plt.ylabel('가격')
plt.show()
```
